```python
On the filtered data, calculate the correlation of 'CO2 emissions (metric tons per capita)' and gdpPercap. What is the correlation and associated p value?
```

```python
# build covariance function calculating the dot product
from typing import List
import numpy as np

def mean(array: List[float]) -> float:
    return sum(array)/len(array)
    
# calculate the difference between each of the values in the array compared to the mean
def dispersion_mean(array: List[float]) -> List[float]:
    mean_value = mean(array)
    
    return [i - mean_value for i in array]
    
def covariance(x: List[float], y: List[float]) -> float:
    # make sure that list x and list y are the same lengtth
    assert len(x) == len(y)
    
    return np.dot(dispersion_mean(x), dispersion_mean(y)) / (len(x) - 1) # using pop-1 rather than pop

def standard_deviation(array: List[float]) -> float:
    mean_dispersion = dispersion_mean(array)
    return np.sqrt(sum([x**2 for x in mean_dispersion])/len(array))

def correlation(x: List[float], y: List[float]) -> float:

    vol_x = standard_deviation(x)
    vol_y = standard_deviation(y)

    return covariance(x, y) / vol_x / vol_y 

gapminder_df_1962_drop_na = gapminder_df_1962.dropna(subset=["CO2 emissions (metric tons per capita)", "gdpPercap"], thresh=2)
correlation_value = correlation(list(gapminder_df_1962["CO2 emissions (metric tons per capita)"]), list(gapminder_df_1962["gdpPercap"]))

# p-value look up from https://www.statisticssolutions.com/free-resources/directory-of-statistical-analyses/pearsons-correlation-coefficient/table-of-critical-values-pearson-correlation/
print(f"corr value: {correlation_value} with pop {len(gapminder_df_1962)} and p-value < 0.05")
```
